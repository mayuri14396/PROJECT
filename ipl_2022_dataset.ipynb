```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
ipl=pd.read_csv("ipl_2022_dataset.csv")
```

```python
ipl.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
ipl.head()
```

```python
ipl.isnull().sum()
```

```python
ipl[ipl['Cost IN $ (000)'].isnull()]
```

```python
ipl['Cost IN $ (000)']=ipl['Cost IN $ (000)'].fillna(0)
```

```python
ipl['COST IN ₹ (CR.)']=ipl['COST IN ₹ (CR.)'].fillna(0)
```

```python
ipl[ipl['2021 Squad'].isnull()]
```

```python
#replace null values presnent in the not partisipated
ipl['2021 Squad']=ipl['2021 Squad'].fillna('not participatd')
```

```python
ipl.isnull().sum()
```

```python
teams=ipl[ipl['COST IN ₹ (CR.)']>0]['Team'].unique()
```

```python
teams
```

```python
ipl['status']=ipl['Team'].replace(teams,'sold')
```

```python
ipl
```

```python
ipl[ipl['Player'].duplicated(keep=False)]
```

```python
# how many players in participated in 2022 ipl auction
ipl.shape[0]
```

```python
# how many players have participated
types=ipl['TYPE'].value_counts()
types.reset_index()
```

```python
plt.pie(types.values,labels=types.index,labeldistance=1.5,autopct='%1.2f%%',shadow=True,startangle=60)
plt.title('Role of Players Participated',fontsize=20)
plt.plot()
```

```python
# player sold and unsold using a bar graph
plt.xlabel('sold and unsold')
plt.ylabel('Number of Players')
plt.title('sold v/s unsold')
plt.plot()plt.figure(figsize=(10,7))
fig=sns.countplot(ipl['status'],palette=['pink','blue'])
```

```python
# player sold and unsold using a bar graph
plt.figure(figsize=(10,7))
fig=sns.countplot(ipl['status'],palette=['pink','blue'])
plt.xlabel('sold and unsold')
plt.ylabel('Number of Players')
plt.title('sold v/s unsold')
plt.plot()

for p in fig.patches:
    fig.annotate(format(p.get_height(),'.0f'),
                 (p.get_x()+p.get_width()/2.,p.get_height()),
                 ha='center',va='center',
                 xytext=(0,4),textcoords='offset points')
```

```python
ipl.groupby('status')['Player'].count()
```

```python
# player sold and unsold using a bar graph
plt.figure(figsize=(10,7))
fig=sns.countplot(ipl[ipl['Team']!='Unsold']['Team'])
plt.xlabel('Team Names')
plt.ylabel('Number of Players')
plt.title('Players Bought by each team',fontsize=15)
plt.plot()


for p in fig.patches:
    fig.annotate(format(p.get_height(),'.0f'),
                 (p.get_x()+p.get_width()/2.,p.get_height()),
                 ha='center',va='center',
                 xytext=(0,4),textcoords='offset points')
```

```python
ipl['retention']=ipl['Base Price']
```

```python
ipl['retention'].replace(['2 Cr','40 Lakh','20 Lakh','1 Cr','75 Lakh','50 Lakh','30 Lakh','1.5 Cr'],'From Auction',inplace=True)
```

```python
# treating base price
ipl['Base Price'].replace('Draft Pick',0,inplace=True)
```

```python
ipl['base_price_unit']=ipl['Base Price'].apply(lambda x: str(x).split(' ')[-1])
ipl['base_price']=ipl['Base Price'].apply(lambda x: str(x).split(' ')[0])
```

```python
ipl['base_price'].replace('Ratined',0,inplace=True)
```

```python
ipl.head()
```

```python
# total players retained and bought
ipl.groupby(['Team','retention'])['retention'].count()[:-1]
```

```python
# player sold and unsold using a bar graph
plt.figure(figsize=(10,7))
fig=sns.countplot(ipl[ipl['Team']!='Unsold']['Team'],hue=ipl['TYPE'])
plt.xlabel('Team Names')
plt.ylabel('Number of Players')
plt.title('Players Bought by each team',fontsize=15)
plt.xticks(rotation=50)
plt.plot()


for p in fig.patches:
    fig.annotate(format(p.get_height(),'.0f'),
                 (p.get_x()+p.get_width()/2.,p.get_height()),
                 ha='center',va='center',
                 xytext=(0,4),textcoords='offset points')
```

```python
# highest amount spent on a single player by each team
ipl[ipl['retention']=='From Auction'].groupby(['Team'])
```

```python
# highest amount spent on a single player by each team
ipl[ipl['retention']=='From Auction'].groupby(['Team'])['COST IN ₹ (CR.)'].max()[:-1].sort_values(ascending=False)
```

```python
# player retained at maximum price.
ipl[ipl['retention']=='Retained'].sort_values(by='COST IN ₹ (CR.)',ascending=False).head(2)
```

```python
ipl[(ipl['retention']=='From Auction') & (ipl['TYPE']=='BATTER')].sort_values(by='COST IN ₹ (CR.)',ascending=False).head()
```

```python
ipl[(ipl['retention']=='From Auction') & (ipl['TYPE']=='BOWLER')].sort_values(by='COST IN ₹ (CR.)',ascending=False).head()
```

```python
ipl[(ipl['retention']=='From Auction') & (ipl['TYPE']=='ALL-ROUNDER')].sort_values(by='COST IN ₹ (CR.)',ascending=False).head()
```
